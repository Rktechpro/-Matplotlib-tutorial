# Bar matplotlib

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[90,60,70,80]
plt.xlabel("Language",fontsize=20)
plt.ylabel("Number",fontsize=20)
plt.title("RkCode", fontsize=20)
g=["y","b","g","r"]
plt.bar(x,y,width=0.4,color=g,align="center",edgecolor="w",linewidth="3",linestyle=":",alpha=0.7,label="first")
plt.bar(x,y,width=0.4,color=g,align="center",edgecolor="w",linewidth="3",linestyle=":",alpha=0.7,label="second")
plt.legend()
plt.show()
```

# ColorMap
The Matplotlib module has a number of available colormaps.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()

plt.show()
```

# Combine Color Size and Alpha
You can combine a colormap with different sizes on the dots. This is best visualized if the dots are transparent:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 15 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

# Matplotlib Pie Charts
With Pyplot, you can use the pie() function to draw pie charts:-

```python
import matplotlib.pyplot as plt
import numpy as np

y=np.array([35,25,25,30])
plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([40,25,25,30])
rklabel=["A","B","C","D"]
plt.pie(y,labels=rklabel)
plt.show()
```

# Start Angle
As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([40,25,25,30])
rklabel=["A","B","C","D"]
plt.pie(y,labels=rklabel,startangle = 90)
plt.show()
```

# Explode
Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([40,25,25,30])
rklabel=["A","B","C","D"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y,labels=rklabel,explode=myexplode)
plt.show()
```

# Shadow
Add a shadow to the pie chart by setting the shadows parameter to True:

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([40,25,25,30])
rklabel=["P","Q","M","N"]
myexplode = [0, 0, 0, 0.2]
plt.pie(y,labels=rklabel,explode=myexplode, shadow = True) 
plt.show()
```

# Colors
You can set the color of each wedge with the colors parameter.

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([80,50,50,50])
rklabel=["P","Q","M","N"]
mycolor=["r","b","g","hotpink"]
myexplode = [0, 0, 0, 0.2]
plt.pie(y,labels=rklabel, colors = mycolor,shadow= True)

plt.show()
```

# Legend With Header
To add a header to the legend, add the title parameter to the legend function.

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([80,50,50,50])
rklabel=["P","Q","M","N"]
mycolor=["y","k","g","hotpink"]
myexplode = [0, 0, 0, 0.2]
plt.pie(y,labels=rklabel, colors = mycolor,shadow= True)
plt.legend()
plt.show()
```
